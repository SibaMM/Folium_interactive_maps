```python
map_dark = folium.Map(location=[-33.9152209,18.3758852], tiles='stamenterrain', attr='stamenterrain', zoom_start=10)
folium.TileLayer('OpenStreetMap', attr='OpenStreetMap').add_to(map_dark)
folium.LayerControl().add_to(map_dark)
map_dark
```

```python
df_locations_new = df_locations[['Latitude', 'Longitude']]
df_locations_new

location_list = df_locations_new.values.tolist()
location_list_size = len(location_list)
location_list_size
```

```python
for point in range(0, location_list_size):
    folium.Marker(location_list[point]).add_to(map_dark)
    folium.Marker(location_list[point], popup=df_locations['Store'][point]).add_to(map_dark)
    folium.Marker(location_list[point], popup=df_locations['Location'][point], icon=folium.Icon(color='darkblue', icon_color='white', icon='tint', angle=0, prefix='fa')).add_to(map_dark)
map_dark
```

```python
map_dark.save('folium_interactive_map.html')
```
